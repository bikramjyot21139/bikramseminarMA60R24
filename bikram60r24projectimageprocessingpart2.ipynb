Getting the code to work in color is pretty straightforward: we first separate the image into three seperate channels, in this case into red, green, and blue channels.

Alternatively the image could be separated into HSV (hue, saturation, value) channels, which could yield a larger improvement on size if finetuned (perhaps saturation information requires fewer singular values for a clear image, but hue information needs more).

We can then perform SVD on each of the color channels, before adding them back together. Note that np.clip is used in the code below as some of the color channels could have negative pixel values at lower values of n, which creates visual artifacts.

```python
U_R, S_R, Vt_R = np.linalg.svd(R, full_matrices=False)
U_G, S_G, Vt_G = np.linalg.svd(G, full_matrices=False)
U_B, S_B, Vt_B = np.linalg.svd(B, full_matrices=False)

n = 50  
R_compressed = np.matrix(U_R[:, :n]) * np.diag(S_R[:n]) * np.matrix(Vt_R[:n, :])
G_compressed = np.matrix(U_G[:, :n]) * np.diag(S_G[:n]) * np.matrix(Vt_G[:n, :])
B_compressed = np.matrix(U_B[:, :n]) * np.diag(S_B[:n]) * np.matrix(Vt_B[:n, :])

# Combining the compressed channels
compressed_image = cv2.merge([np.clip(R_compressed, 1, 255), np.clip(G_compressed, 1, 255), np.clip(B_compressed, 1, 255)])
compressed_image = compressed_image.astype(np.uint8)
plt.imshow(compressed_image)
plt.title('n = %s' % n)
plt.show()

# Plotting the compressed RGB channels
plt.subplot(1, 3, 1)
plt.imshow(R_compressed, cmap='Reds_r')
plt.subplot(1, 3, 2)
plt.imshow(B_compressed, cmap='Blues_r')
plt.subplot(1, 3, 3)
plt.imshow(G_compressed, cmap='Greens_r')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import numpy as np
import matplotlib.pyplot as plt
import cv2
class NoisyImageDataset(Dataset):
    def __init__(self, image, noise_type='gaussian', sigma=25, prob=0.02):
        self.image = image.astype(np.float32) / 255.0
        self.noise_type = noise_type
        self.sigma = sigma
        self.prob = prob

    def __len__(self):
        return 1  # Only one image

    def __getitem__(self, idx):
        clean = self.image.transpose(2, 0, 1)  # C,H,W
        noisy = self.add_noise(self.image).transpose(2, 0, 1)
        return torch.tensor(noisy, dtype=torch.float32), torch.tensor(clean, dtype=torch.float32)

    def add_noise(self, image):
        if self.noise_type == 'gaussian':
            noise = np.random.normal(0, self.sigma/255.0, image.shape).astype(np.float32)
            noisy = image + noise
        elif self.noise_type == 'salt_pepper':
            noisy = image.copy()
            rnd = np.random.rand(*image.shape[:2])
            noisy[rnd < self.prob/2] = 0
            noisy[rnd > 1 - self.prob/2] = 1
        else:
            noisy = image
        return np.clip(noisy, 0.0, 1.0)


# ---------- Convolutional Autoencoder ----------
class DenoiseAutoencoder(nn.Module):
    def __init__(self):
        super(DenoiseAutoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(),
            nn.Conv2d(64, 32, 3, padding=1), nn.ReLU(),
            nn.Conv2d(32, 16, 3, padding=1), nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Conv2d(16, 32, 3, padding=1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.Conv2d(64, 3, 3, padding=1), nn.Sigmoid()
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x


# ---------- Load Image ----------
img = cv2.imread("/kaggle/input/inputtrials/datasetrial1.png")  # Replace with your image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = img.astype(np.float32)

# ---------- Dataset and DataLoader ----------
dataset = NoisyImageDataset(img, noise_type='gaussian', sigma=25)
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

# ---------- Training ----------
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = DenoiseAutoencoder().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

epochs = 200
for epoch in range(epochs):
    for noisy, clean in dataloader:
        noisy, clean = noisy.to(device), clean.to(device)
        optimizer.zero_grad()
        outputs = model(noisy)
        loss = criterion(outputs, clean)
        loss.backward()
        optimizer.step()
    if (epoch+1) % 50 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.6f}")
model.eval()
with torch.no_grad():
    noisy_img, clean_img = dataset[0]
    noisy_tensor = noisy_img.unsqueeze(0).to(device)
    denoised_tensor = model(noisy_tensor).cpu().squeeze(0).numpy()
    denoised_img = np.clip(denoised_tensor.transpose(1, 2, 0)*255, 0, 255).astype(np.uint8)
    noisy_img_display = np.clip(noisy_img.transpose(1, 2, 0)*255, 0, 255).astype(np.uint8)

# ---------- Display ----------
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(img.astype(np.uint8)); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(noisy_img_display); plt.title("Noisy"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(denoised_img); plt.title("Denoised (NN)"); plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import svd
from scipy.ndimage import median_filter
import cv2

# ---------- Gaussian Noise Denoising ----------
def svd_denoise_gaussian(image, k):
    img = image.astype(float)
    denoised = np.zeros_like(img)

    if img.ndim == 2:  # grayscale
        U, s, Vt = svd(img, full_matrices=False)
        S = np.diag(s[:k])
        denoised = U[:, :k] @ S @ Vt[:k, :]
    else:  # color
        for c in range(img.shape[2]):
            U, s, Vt = svd(img[:, :, c], full_matrices=False)
            S = np.diag(s[:k])
            denoised[:, :, c] = U[:, :k] @ S @ Vt[:k, :]

    return np.clip(denoised, 0, 255).astype(np.uint8)


# ---------- Salt & Pepper Noise Denoising ----------
def svd_denoise_sp(image, k, median_size=3):
    if image.ndim == 3:  # color
        med = np.zeros_like(image)
        for c in range(image.shape[2]):
            med[:, :, c] = median_filter(image[:, :, c], size=median_size)
    else:  # grayscale
        med = median_filter(image, size=median_size)

    med = med.astype(float)
    denoised = np.zeros_like(med)

    if med.ndim == 2:
        U, s, Vt = svd(med, full_matrices=False)
        S = np.diag(s[:k])
        denoised = U[:, :k] @ S @ Vt[:k, :]
    else:
        for c in range(med.shape[2]):
            U, s, Vt = svd(med[:, :, c], full_matrices=False)
            S = np.diag(s[:k])
            denoised[:, :, c] = U[:, :k] @ S @ Vt[:k, :]

    return np.clip(denoised, 0, 255).astype(np.uint8)


# ---------- Example Usage ----------
def add_gaussian_noise(image, mean=0, sigma=25):
    gauss = np.random.normal(mean, sigma, image.shape).astype(np.float32)
    noisy = image.astype(np.float32) + gauss
    return np.clip(noisy, 0, 255).astype(np.uint8)

def add_sp_noise(image, prob=0.02):
    noisy = image.copy()
    rnd = np.random.rand(*image.shape[:2])
    noisy[rnd < prob/2] = 0
    noisy[rnd > 1 - prob/2] = 255
    return noisy

# Load image (change path to your image)
img = cv2.imread("/kaggle/input/inputtrials/datasetrial1.png")  # <-- replace with your image path
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Add noise
noisy_gauss = add_gaussian_noise(img)
noisy_sp = add_sp_noise(img)

# Apply denoising
denoised_gauss = svd_denoise_gaussian(noisy_gauss, k=50)
denoised_sp = svd_denoise_sp(noisy_sp, k=50, median_size=3)

# Plot results
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes[0, 0].imshow(img); axes[0, 0].set_title("Original"); axes[0, 0].axis("off")
axes[0, 1].imshow(noisy_gauss); axes[0, 1].set_title("Gaussian Noise"); axes[0, 1].axis("off")
axes[0, 2].imshow(denoised_gauss); axes[0, 2].set_title("SVD Denoised (Gaussian)"); axes[0, 2].axis("off")

axes[1, 0].imshow(img); axes[1, 0].set_title("Original"); axes[1, 0].axis("off")
axes[1, 1].imshow(noisy_sp); axes[1, 1].set_title("CNN_fliterednoise"); axes[1, 1].axis("off")
axes[1, 2].imshow(denoised_sp); axes[1, 2].set_title("CNNDenoised"); axes[1, 2].axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from skimage import io, color, util
from skimage.metrics import peak_signal_noise_ratio as psnr

# Load image and convert to grayscale
X = color.rgb2gray(io.imread('/kaggle/input/inputtrials/datasetrial1.png'))

# Add Gaussian noise
X_noisy = util.random_noise(X, mode='gaussian', var=0.01)

# Perform SVD
U, S, VT = np.linalg.svd(X_noisy, full_matrices=False)

# Keep top k singular values for reconstruction
k = 50
S_k = np.diag(S[:k])
X_reconstructed = U[:, :k] @ S_k @ VT[:k, :]

# Compute PSNR as accuracy metric
accuracy_svd = psnr(X, X_reconstructed)
print("SVD Reconstruction PSNR:", accuracy_svd)
```

```python
import numpy as np
from skimage import io, color, util
from skimage.metrics import peak_signal_noise_ratio as psnr
import torch
import torch.nn as nn
import torch.nn.functional as F

# Load image and convert to grayscale
X = color.rgb2gray(io.imread('/kaggle/input/inputtrials/datasetrial1.png'))

# Normalize to [0,1] and add batch/channel dimensions
X_tensor = torch.tensor(X, dtype=torch.float32).unsqueeze(0).unsqueeze(0)

# Add Gaussian noise
noise = torch.randn_like(X_tensor) * 0.1
X_noisy_tensor = X_tensor + noise
X_noisy_tensor = torch.clamp(X_noisy_tensor, 0., 1.)

# Simple CNN denoiser
class DenoiseCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 64, 3, padding=1)
        self.conv2 = nn.Conv2d(64, 64, 3, padding=1)
        self.conv3 = nn.Conv2d(64, 1, 3, padding=1)
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = torch.sigmoid(self.conv3(x))
        return x

model = DenoiseCNN()

# Simple training loop (small number of steps for demonstration)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.MSELoss()

for _ in range(200):
    optimizer.zero_grad()
    output = model(X_noisy_tensor)
    loss = criterion(output, X_tensor)
    loss.backward()
    optimizer.step()

# Get denoised output
X_denoised = model(X_noisy_tensor).detach().numpy().squeeze()

# Compute PSNR
accuracy_cnn = psnr(X, X_denoised)
print("CNN Denoising PSNR:", accuracy_cnn)
```
